```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

# Initialize final DataFrame
final1 = pd.DataFrame()

# Loop through pages
for j in range(1, 3):
    url = f"https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={j}"

    # Set headers to mimic your browser request
    headers = {
        "accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7",
        "accept-encoding": "gzip, deflate, br, zstd",
        "accept-language": "en-GB,en;q=0.9,mr-IN;q=0.8,mr;q=0.7,en-US;q=0.6,fr;q=0.5",
        "cache-control": "max-age=0",
        "cookie": "HOWTORT=ul=1730788998778&r=https%3A%2F%2Fwww.ambitionbox.com%2F",  # Replace if this cookie expires
        "dnt": "1",
        "if-none-match": '"369bb-Lmsq3bp/iW/qg2+q/XsxGYTyfFk"',
        "priority": "u=0, i",
        "sec-ch-ua": '"Chromium";v="130", "Google Chrome";v="130", "Not?A_Brand";v="99"',
        "sec-ch-ua-mobile": "?0",
        "sec-ch-ua-platform": '"Linux"',
        "sec-fetch-dest": "document",
        "sec-fetch-mode": "navigate",
        "sec-fetch-site": "same-origin",
        "sec-fetch-user": "?1",
        "upgrade-insecure-requests": "1",
        "user-agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0 Safari/537.36"
    }
    # Send a GET request with headers
    response = requests.get(url, headers=headers, timeout=10)
    webpage = response.text
    outer_soup = BeautifulSoup(webpage, 'lxml')
    company = outer_soup.find_all('div', class_="companyCardWrapper")

    # Lists to collect data
    name = []
    rating = []
    reviews = []
    company_types = []
    
    # Extract data for each company
    for i in company:
        name.append(i.find('h2').text.strip())
        rating.append(i.find("span", class_="companyCardWrapper__companyRatingValue").text if i.find("span", class_="companyCardWrapper__companyRatingValue") else None)

        x = i.find_all('div', class_="companyCardWrapper__tertiaryInformation")
        for t in x:
            text = t.text
            review = re.search(r"(\d+\.?\d*k?)", text)
            reviews.append(review.group(1) if review else 0)

        for line in i.find_all('div', class_="companyCardWrapper__interLinkingWrapper"):
            text = line.text.strip()
            x = text.split('|')
            company_types.append(x[2].strip() if len(x) > 2 else None)

    # Create a DataFrame for the current page
    df = pd.DataFrame({'name': name, 'rating': rating, 'reviews': reviews, 'type': company_types})

    # Concatenate to the final DataFrame
    final1 = pd.concat([final1, df], ignore_index=True)
```
